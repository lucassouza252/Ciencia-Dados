```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
num_linhas = data.shape[0]
data.shape
```

```python
# Data Frame for study null values
nulls = pd.DataFrame()
nulls["Colunas"] = data.columns

# Nulls by columns
data_null = data.isnull().sum()
null_value = []
for x in data_null:
    null_value.append(x)
nulls["nulls"] = null_value

# Nulls percentage
nulls["nulls_percent"] = nulls["nulls"]/num_linhas
```

```python
# Show Nulls Values Percentage > 0 Descending
nulls[nulls["nulls_percent"] > 0].sort_values(by="nulls_percent", ascending=False)
```
