#### By Sex

```python
g = sns.FacetGrid(data, col="sex")
g.map(sns.histplot, "age")
```

```python
g = sns.FacetGrid(data, col='sex')
g.map(sns.histplot, 'ejection_fraction')
```

```python
g = sns.FacetGrid(data, col='sex')
g.map(sns.histplot, 'creatinine_phosphokinase')
```

```python
g = sns.FacetGrid(data, col='sex', hue='smoking')
g.map(sns.histplot, 'serum_creatinine')
```

```python
g = sns.FacetGrid(data, col='sex', hue='smoking')
g.map(sns.histplot, 'serum_sodium')
g.add_legend()
```

```python
g = sns.FacetGrid(data, col='sex', hue='smoking')
g.map(sns.histplot, 'platelets')
g.add_legend()
```
